```python
def apply_key_message_binning(row):
  if row['km_block_counts'] == 0:
    row['km_reached'] = 0
    row['km_bin'] = '0'
  else:
    row['km_reached'] = 1
    if row['km_block_counts'] == 1:
      row['km_bin'] = '1'
    elif (row['km_block_counts'] > 1) & (row['km_block_counts'] <= 4):
      row['km_bin'] = '2-4'
    else:
      row['km_bin'] = '>4'

  return row
```

```python
df_joined = df.merge(block_counts_by_call_id, how='left')

df2 = df_joined.apply(lambda row: apply_key_message_binning(row), axis = 1)
```

There are quite a lot of outliers, while these are very important to understand if something went wrong with these calls causing adverse caller impact, for the first round of analysis, we will check what happens when call duration is not an outlier - where we have defined outlier as when call duration is greater than the upper fence of the violin distribution (1.5 times IQR )for each km_bin.

For 0 KMs reached, we label outliers as calls with call duration > 136 seconds (Derived from values of upper fence of Violin plot above).

For 1 KMs reached, we label outliers as calls with call duration > 365 seconds.

For 2-4 KMs reached, we label outliers as calls with call duration > 1470 seconds.

For >4 KMs reached, we label outliers as calls with call duration > 2158 seconds

```python
def apply_call_duration_outlier_checks(row):
  row['call_duration_outlier'] = 0
  if row['km_bin'] == '0':
    if row['call_duration'] > 136:
      row['call_duration_outlier'] = 1
  else:
    if row['km_bin'] == '1':
      if row['call_duration'] > 365:
        row['call_duration_outlier'] = 1
    elif row['km_bin'] == '2-4':
      if row['call_duration'] > 1470:
        row['call_duration_outlier'] = 1
    elif row['km_bin'] == '>4':
      if row['call_duration'] > 2158:
        row['call_duration_outlier'] = 1
    else:
        row['call_duration_outlier'] = 0
  return row

df_calls = df_calls.apply(lambda row: apply_call_duration_outlier_checks(row), axis = 1)
df_calls.groupby(['call_duration_outlier']).size()
```
