```python
query = """
SELECT EXTRACT (YEAR FROM call_date) as Year,
    EXTRACT (MONTH FROM call_date) as Month,
    gender,
    age,
    case  
    when rural_or_urban like '%town%' then 'Town'
    when rural_or_urban like '%village%' then 'Village'
    when rural_or_urban like '%city%' then 'City'
    when rural_or_urban like '%no_answer%' then 'No answer'
    when rural_or_urban like '%wish%' then 'did not wish to answer'
    else 'Other' 
    END as urban_rural
    ,
    subscriber_pereferred_language,
    count(distinct subscriber_id) Subscribers
    FROM `viamo-datakind.datadive.321_sessions_1122` 
where gender is not null and 
    age is not null and 
    organization_country='Nigeria' and
    char_length(rural_or_urban) > 0
group by 1,
    2,
    3,
    4,
    5,
    6
"""

df_n_urbanrural = pd.read_gbq(query)
```
